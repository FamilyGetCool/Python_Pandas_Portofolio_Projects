### 20. Classify Movies Based on Ratings (Excellent, Good, Average, and Below Average)


```python
df['Rating'].describe()
```

```python
def rating(x):
    if x >= 7.5:
        return 'Excellent'
    if x >= 7:
        return 'Good'
    if x >= 6.5:
        return 'Average'
    else:
        return 'Below Average'

df['Comment'] = df['Rating'].apply(rating)
df.head(3)
```

### 22. Find Unique Values From Genre 


```python
# Splitting values by comma twice to separate all values 
List1 = [] 
List2 = []
for x in df['Genre']:
    List1.append(x.split(',')) # Separate the values with comma: ['a','b','c'],['c','b','d']
for y in List1:
    for z in y:
        List2.append(z.split(',')) # Separate the values with comma again: ['a'],['b'],['c'],['c'],['b'],['d']
# Create an unique_list and start appending unique values to it by using "not in".
Unique_List = []
for values in List2:
    if values not in Unique_List:
        Unique_List.append(values)
Unique_List
```

```python
print('Number of Genres:', len(Unique_List))
```
