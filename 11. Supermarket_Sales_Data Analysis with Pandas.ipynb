```python
#                                    Detail
#1. Invoice id: Computer generated sales slip invoice identification number
#2. Branch: Branch of supercenter (3 branches are available identified by A, B and C).
#3. City: Location of supercenters
#4. Customer type: Type of customers, recorded by Members for customers using member card and Normal for without member card.
#5. Gender: Gender type of customer
#6. Product line: General item categorization groups - Electronic accessories, Fashion accessories, Food and beverages, Health and beauty, Home and lifestyle, Sports and travel
#7. Unit price: Price of each product in $
#8. Quantity: Number of products purchased by customer
#9. Tax: 5% tax fee for customer buying
#10. Total: Total price including tax
#11. Date: Date of purchase (Record available from January 2019 to March 2019)
#12. Time: Purchase time (10am to 9pm)
#13. Payment: Payment used by customer for purchase (3 methods are available â€“ Cash, Credit card and Ewallet)
#14. COGS: Cost of goods sold
#15. Gross margin percentage: Gross margin percentage
#16. Gross income: Gross income
#17. Rating: Customer stratification rating on their overall shopping experience (On a scale of 1 to 10)
```

### 9. Display Daily Sales by Day of the Week

```python
dayofweek = {
    0:'MON',
    1:'TUE',
    2:'WED',
    3:'THUR',
    4:'FRI',
    5:'SAT',
    6:'SUN'
}
```

```python
df['Day_of_week'] = df['Date'].dt.day_of_week.map(dayofweek)
df.head(2)
```

```python
test7 = df.groupby('Day_of_week')['Total'].sum().reset_index()
dow_categories = ['MON','TUE','WED','THUR','FRI','SAT','SUN']
test7['Day_of_week'] = pd.Categorical(test7['Day_of_week'], categories = dow_categories)
test7.sort_values(by = 'Day_of_week')
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Day_of_week', y='Total', data=test7)
plt.title('Daily Sales by Day of Week', fontsize=17)
plt.xlabel('Day of Week', fontsize=17)
plt.ylabel('Total Sales', fontsize=17)
plt.show()
```

### 10. Find the Highest Monthly Sales 

#### Method 1

```python
df['Month'].replace({
    1:"January", 
    2:"February", 
    3:"March"
}, inplace=True)
df.head(2)
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Month', y='Total', data=df)
plt.title('Total Sales by Month', fontsize=17)
plt.xlabel('Month', fontsize=17)
plt.ylabel('Total Sales', fontsize=17)
plt.show()
```

#### Method 2

```python
test8 = df.groupby('Month')['Total'].sum().reset_index()
months_categories1 = ['January', 'February', 'March']
test8['Month'] = pd.Categorical(test8['Month'], categories=months_categories1)
test8.sort_values('Month')
```

```python
plt.figure(figsize=(10,5))
sns.barplot(x='Month', y='Total', data=test8)
plt.title('Total Sales by Month', fontsize=17)
plt.xlabel('Month', fontsize=17)
plt.ylabel('Total Sales', fontsize=17)
plt.show()
```
