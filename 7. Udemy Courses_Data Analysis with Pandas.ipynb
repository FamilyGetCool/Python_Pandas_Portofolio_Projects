### 7. Find Out Number of Courses per Subject


```python
# Method 1
df['subject'].value_counts().reset_index()
```

```python
# Method 2
test2 = df.groupby('subject')['course_id'].nunique().sort_values(ascending=False).reset_index()
test2
```

```python
sns.barplot(x='subject', y='course_id', data=test2)
plt.xlabel('Subject', fontsize=12)
plt.ylabel('Number of Courses', fontsize=12)
plt.xticks(rotation=65)
```

```python
sns.countplot(x='subject', data=df)
plt.xlabel('Subject', fontsize=12)
plt.ylabel('Number of Courses', fontsize=12)
plt.xticks(rotation=65)
```

### 8. How Many Different Levels of Courses Does Udemy Course Provide?

```python
sns.barplot(x='index', y='level', data=test3)
plt.xlabel('level', fontsize=12)
plt.ylabel('Number of Courses', fontsize=12)
plt.xticks(rotation=65)
```

```python
sns.countplot(x='level', data=df)
plt.xlabel('level', fontsize=12)
plt.ylabel('Number of Courses', fontsize=12)
plt.xticks(rotation=65)
```

### 9. Display the Count of Paid and Free Courses 


```python
df['is_paid'].replace({True: 'Paid', False: 'Free'}, inplace=True)
df['is_paid'].value_counts()
```

```python
sns.countplot(x='is_paid', data=df)
plt.xlabel('Type', fontsize=12)
plt.ylabel('Number of Courses', fontsize=12)
```

### 10. Which Type of Courses Have More Lectures (Free or Paid)?


```python
test4 = df.groupby('is_paid')['num_lectures'].sum().sort_values(ascending=False).reset_index()
test4
```

```python
sns.barplot(x='is_paid', y='num_lectures', data=test4)
plt.xlabel('Type', fontsize=12)
plt.ylabel('Number of Lectures', fontsize=12)
```

### 11. Which Type of Courses Have More Subscribers (Free or Paid)?


```python
test5 = df.groupby('is_paid')['num_subscribers'].sum().sort_values(ascending=False).reset_index()
test5
```

```python
sns.barplot(x='is_paid', y='num_subscribers', data=test5)
plt.xlabel('Type', fontsize=12)
plt.ylabel('Num of Subscribers', fontsize=12)
```

### 12. Which Level Has the Highest Number of Subscribers?


```python
test6 = df.groupby('level')['num_subscribers'].sum().sort_values(ascending=False).reset_index()
test6
```

```python
sns.barplot(x='level', y='num_subscribers', data=test6)
plt.xlabel('Level', fontsize=12)
plt.ylabel('Num of Subscribers', fontsize=12)
```
